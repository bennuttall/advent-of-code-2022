```python
from itertools import cycle

def get_jets_and_rocks():
    jets = cycle(">>><<><>><<<>><>>><<<>>><<<><<<>><>><<>>")

    rocks = cycle((
        {(0, 0), (1, 0), (2, 0), (3, 0)},

        {        (1, 0),
         (0, 1), (1, 1), (2, 1),
                 (1, 2)},

        {                (2, 0),
                         (2, 1),
         (0, 2), (1, 2), (2, 2)},

        {(0, 0), (1, 0),
         (0, 1), (1, 1)}
    ))
    
    return jets, rocks
```

```python
def apply_jet(rock):
    jet = next(jets)
    min_x = min(x for (x, y) in rock)
    max_x = max(x for (x, y) in rock)
    if jet == '<' and min_x > 0:
        dx = -1
    elif jet == '>' and max_x < 6:
        dx = 1
    else:
        return rock
    new_rock = {(rx + dx, ry) for (rx, ry) in rock}
    if len(new_rock & grid) == 0:
        return new_rock
    return rock
```

```python
def apply_fall(rock):
    min_y = min(y for (x, y) in rock)
    if min_y > 0:
        new_rock = {(x, y - 1) for (x, y) in rock}
        if len(new_rock & grid) == 0:
            return new_rock
    return rock
```

```python
def rock_fall(grid):
    rock = position_rock(next(rocks))
    while True:
        grid -= rock
        rock_after_jet = apply_jet(rock)
        rock_after_fall = apply_fall(rock_after_jet)
        grid |= rock_after_fall
        draw_grid()
        if rock_after_fall == rock_after_jet:
            draw_grid()
            return
        rock = rock_after_fall
        draw_grid()
```

```python
def draw_grid():
    max_y = max(y for (x, y) in grid)
    for y in range(max([max_y, 7]), -1, -1):
        for x in range(7):
            char = '#' if (x, y) in grid else '.'
            print(char, end='')
        print()
    print('=' * 7)
    print()
```

```python
grid = set()
jets, rocks = get_jets_and_rocks()

for i in range(2):
    rock_fall(grid)
```
